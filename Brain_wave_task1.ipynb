```python
# Calculate profit for each product
data["profit"] = data["retail_price"] - data["discounted_price"]

# Calculate profit percentage
data["profit_percentage"] = (data["profit"] / data["retail_price"]) * 100

# Display the top 10 products with the highest profit
top_profit_products = data.sort_values(by="profit", ascending=False).head(10)
print(top_profit_products[["product_name", "retail_price", "discounted_price", "profit", "profit_percentage"]])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Categorize products as "Discounted" or "Non-Discounted"
data["discount_applied"] = data["discount_percentage"].apply(lambda x: "Yes" if x > 0 else "No")
# Sales trend comparison
plt.figure(figsize=(8, 5))
sns.boxplot(x="discount_applied", y="retail_price", data=data, palette="coolwarm")

# Labels
plt.xlabel("Discount Applied")
plt.ylabel("Retail Price")
plt.title("Sales Price Distribution: Discounted vs. Non-Discounted Products")

plt.show()
```

```python
# Convert timestamp column to datetime format
data["crawl_timestamp"] = pd.to_datetime(data["crawl_timestamp"])

# Extract month and year
data["Month"] = data["crawl_timestamp"].dt.month
data["Year"] = data["crawl_timestamp"].dt.year

# Aggregate sales by month
monthly_sales = data.groupby(["Year", "Month"])["discounted_price"].sum().reset_index()

# Plot sales trend
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales["discounted_price"], marker="o")

plt.xlabel("Time (Year-Month)")
plt.ylabel("Total Sales Revenue")
plt.title("Sales Trend Over Time")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by brand and calculate average rating and total sales
brand_analysis = data.groupby("brand").agg({
    "overall_rating": "mean",
    "Sales_Volume": "sum",
    "product_name": "count"
}).reset_index()


brand_analysis = brand_analysis.sort_values(by="overall_rating", ascending=False)


plt.figure(figsize=(12, 6))
sns.scatterplot(
    x=brand_analysis["overall_rating"],
    y=brand_analysis["Sales_Volume"],
    size=brand_analysis["product_name"],
    sizes=(50, 500),
    hue=brand_analysis["overall_rating"],
    palette="coolwarm",
    alpha=0.8
)


plt.xlabel("Average Brand Rating")
plt.ylabel("Total Sales Volume")
plt.title("Brand Rating vs Sales Volume (Bubble Size = No. of Products)")

plt.show()
```

knowing which season as high discount price?

```python
# Aggregate total sales per season
seasonal_sales = data.groupby("Season")["discounted_price"].sum().reset_index()

# Sort seasons in order
season_order = ["Winter", "Spring", "Summer", "Autumn"]
seasonal_sales["Season"] = pd.Categorical(seasonal_sales["Season"], categories=season_order, ordered=True)

# Sort DataFrame
seasonal_sales = seasonal_sales.sort_values("Season")

# Display the result
print(seasonal_sales)
```

Conclusion: Winter sees the highest discounted sales, indicating that customers are more likely to buy during seasonal discounts, while Summer has the lowest, suggesting lower demand or fewer discounts.

Top 5 Best-Selling Products Across All Seasons

```python
# Ensure 'Sales_Volume' exists or is calculated
if "Sales_Volume" not in data.columns:
    data["Sales_Volume"] = data["profit"] / (data["retail_price"] - data["discounted_price"])

# Group by product and sum total sales
top_products = data.groupby("product_name")["Sales_Volume"].sum().reset_index()

# Sort by highest sales and take the top 5
top_5_products = top_products.sort_values(by="Sales_Volume", ascending=False).head(5)

# Display top 5 selling products
print(top_5_products)

# Plot bar chart for top 5 selling products
plt.figure(figsize=(10,5))
sns.barplot(x="Sales_Volume", y="product_name", data=top_5_products, palette="viridis")
plt.xlabel("Total Sales Volume")
plt.ylabel("Product Name")
plt.title("Top 5 Best-Selling Products Across All Seasons")
plt.show()
```

Conclusion: The top-selling products include tech accessories, fashion items, and home decor, suggesting that protective gear, apparel, and aesthetic enhancements drive consistent sales across all seasons.

Sales Revenue vs. Discount Percentage and Impact of Discount Levels on Sales?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Categorize discount levels
bins = [0, 20, 50, 100]  # Discount percentage ranges
labels = ["Low Discount (0-20%)", "Medium Discount (21-50%)", "High Discount (51-100%)"]
data["Discount_Level"] = pd.cut(data["discount_percentage"], bins=bins, labels=labels)

# Aggregate sales per discount level
discount_sales = data.groupby("Discount_Level")["discounted_price"].sum().reset_index()

# Scatter Plot: Discount Percentage vs. Sales Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data["discount_percentage"], y=data["discounted_price"], alpha=0.6, color="blue")
plt.xlabel("Discount Percentage")
plt.ylabel("Sales Revenue")
plt.title("Sales Revenue vs. Discount Percentage")
plt.show()

# Bar Plot: Sales Trend Based on Discount Levels
plt.figure(figsize=(8, 5))
sns.barplot(x="Discount_Level", y="discounted_price", data=discount_sales, palette="coolwarm")
plt.xlabel("Discount Level")
plt.ylabel("Total Sales Revenue")
plt.title("Impact of Discount Levels on Sales")
plt.show()
```

Conclusion: Products with median discount levels tend to achieve higher sales, indicating an optimal balance between price reduction and perceived value.
